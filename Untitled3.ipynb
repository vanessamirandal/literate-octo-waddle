```python
metric ='wait_time_min'
q1=df[metric].quantile(0.25)
q3=df[metric].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
df['outlier_wait_time_min'] = (df[metric]< lower_bound) | (df[metric]> upper_bound)
```

```python
# Detractor 0-6
# Passive 7-8
# Promoter 9-12
```

```python
def classify_satisfaction(x):
  if x < 7:
    cat='Detractor 0-6'
  elif x >= 7 and x <= 8:
    cat='Passive'
  elif x >= 8:
    cat='Promoter'
  else:
    cat='Unknown'
    return cat
```

```python
df['clas_satisfaction_score'] = df['satisfaction_score'].apply(classify_satisfaction)
```
